```python
import re
import pandas as pd
```

```python
f = open('WhatsApp_Chat_with_Major_Project[1].txt','r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print (type(data))
```

```python
pattern = r"(\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}\s(?:am|pm))\s-\s([^:]+:\s.*)"

```

```python
matches = re.findall(pattern, data) 
matches
```

```python
dates = [match[0] for match in matches]
messages = [match[1] for match in matches]
messages
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date' :dates})
# Step 1: make sure it's string
df
```

```python
df['message_date'] = df['message_date'].astype(str)
df
```

```python
df["message_time"] = df["message_date"].str.split(", ").str[1]
df["message_date"] = df["message_date"].str.split(", ").str[0]
df
```

```python
df['message_date'] = pd.to_datetime(df['message_date'],format='mixed')
df
```

```python
df["message_time"] =pd.to_datetime(df['message_time'],format="%I:%M %p").dt.time

df
```

```python
df['year']= df['message_date'].dt.year
```

```python
df['month'] = df['message_date'].dt.month_name()
```

```python
df['day']= df['message_date'].dt.day
```

```python
df['hour'] =df['message_time'].dt.hour
```

```python
df['minute'] = df['message_time'].dt.minute
```
